## Обращаемся к API hh.ru

Документация - https://dev.hh.ru/

```python
url = "https://api.hh.ru/vacancies?text=middle python разработчик&per_page=20"
```

```python
url = "https://api.hh.ru/vacancies"
```

Полный список параметров: https://api.hh.ru/openapi/redoc#tag/Poisk-vakansij/operation/get-vacancies

```python
url_params = {
    "text": "middle python",
    # "search_field": "name",
    # "per_page": "20",
    # "page": "1",
    # "area": "2"
}
```

Запускаем запрос

```python
result = requests.get(url, params=url_params)
```

Просматриваем запрошенный URL

```python
result.url
```

Проверяем ответ

```python
result.status_code
```

Проверяем содержание ответа

```python
result.json()
```

Количество вакансий

```python
len(result.json().get('items'))
```

## Просматриваем список вакансий

```python
vacancies = result.json().get('items')
```

```python
for i, vacancy in enumerate(vacancies):
    print(i + 1, vacancy['name'], vacancy['url'], vacancy['alternate_url'])
```
